```python
import random

class Case:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __eq__(self,other):
        return (self.x == other.x) and (self.y == other.y)
    def __str__(self):
        return "Case: x="+str(self.x)+" , y="+str(self.y)
    def __repr__(self):
        return "("+str(self.x)+", "+str(self.y)+")"
    def adjacentes(self,jeu):
        listeCases = []
        for x in range(self.x-1,self.x+2):              #
            for y in range(self.y-1,self.y+2):
                if x < jeu.x and y < jeu.y and x >= 0 and y >= 0:
                    listeCases.append(Case(x,y))
        listeCases.remove(Case(self.x,self.y))
        return listeCases
                    

class Créature:
    def __init__(self,nom,c):
        self.nom = nom
        self.position = c # c est une Case
    def __str__(self):
        return self.nom+" est sur "+str(self.position)
    def __repr__(self):
        return self.nom+" est sur "+str(self.position)
    def choisirCible(self,jeu):
        #on récupère les cases adjacentes à celle de la créature
        listeCases = self.position.adjacentes(jeu)
        
        #Si une des cases est occupée on la renvoie
        for case in listeCases:
            if jeu.estOccupée(case):       
                return case,True
        
        #Sinon on en renvoie une au hasard
        return random.choice(listeCases),False    


class Jeu:
    def __init__(self, listeDesCréatures, dimensions):
        
        self.listeDesCases = []
        self.fini = False
        
        for i in range(dimensions[0]):
            for j in range(dimensions[1]):
                self.listeDesCases.append(Case(i,j))
                
        self.x = dimensions[0]
        self.y = dimensions[1]
        self.listeDesCréatures = listeDesCréatures
        self.tour = 1
        self.actif = 0 # index du monstre actif dans listeDesCréatures
        
    def __str__(self):
        
        s = ""
        for i in self.listeDesCréatures:
            print(i)
        s = "Tour: "+str(self.tour)+" actif: "+str(self.actif)
        
        
        for i in range(len(self.listeDesCases)):
            if i%self.x == 0:
                s+="\n"
            if self.estOccupée(self.listeDesCases[i]):
                s+= " "+str(self.listeDesCréatures.index(self.estOccupée(self.listeDesCases[i])))+ " "
            else:
                s+= " X " 
        
        return s
    
    def estOccupée(self,c):
        for i in self.listeDesCréatures:
            if i.position == c:
                return i      #Si la case est occupée, on la renvoie, sinon on renvoie simplement False
        return False
    
    def deplacer(self):
        self.listeDesCréatures[self.actif].position, aGagné = self.listeDesCréatures[self.actif].choisirCible(self)
        if aGagné :
            print(self.listeDesCréatures[self.actif].nom+ " a gagné!")
            self.fini = True
        else:
            print("Pas de chance, au tour de "+self.listeDesCréatures[self.actif].nom)
            self.tour +=1
            if self.actif == 0:
                self.actif = 1
            else:
                self.actif = 0
            print(self)
        

coteDamier = 4 # nb de cases pour le coté du damier
cr1 = Créature("Griffon",Case(0,0)) # positionner sur la case 1,1
cr2 = Créature("Pégase",Case(3,3)) # positionner à l'opposé
l =[cr1,cr2]

j = Jeu(l,[coteDamier,coteDamier]) # création de l'objet j de type Jeu

print(j.listeDesCases[7].adjacentes(j))
print(j.listeDesCréatures[0].choisirCible(j))

while j.fini == False:

    j.deplacer()
```
