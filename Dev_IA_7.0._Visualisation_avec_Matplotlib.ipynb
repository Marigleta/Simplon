## 1.	Préparer son jeu de données

```python
import pandas as pd
donnees=pd.read_csv("Jeux_de_donnees/120-years-of-olympic-history-athletes-and-results.csv", index_col=[1])
donnees
```

### Les valeurs manquantes
Il faut savoir que certaines fonctions graphiques ignorent par défaut les individus ou observations contenant des valeurs manquantes, donc sans le savoir, vous travaillez sur un sous-ensemble de vos données. Il existe aussi des fonctions graphiques qui génèrent des erreurs lorsqu’il y a des valeurs manquantes dans la variable qui vous intéresse. Ainsi, il est important de traiter ces valeurs manquantes selon l’objectif de la visualisation.

```python
donnees.isna()
```

```python
donnees.isna().sum()
```

On utilise la méthode fillna() de Pandas sur la colonne Medal de notre DataFrame stocké dans la variable donnees pour remplacer les NaN de cette colonne par la valeur None.

```python
donnees['Medal']=donnees['Medal'].fillna("None")
donnees
```

```python
donnees['Medal']=donnees['Medal'].fillna("None")
donnees_NaN_supprimes = donnees.dropna(how="any")
donnees[['Age','Height', 'Weight']] = donnees[['Age','Height','Weight']].fillna(donnees[['Age','Height','Weight']].mean())
donnees_NaN_remplaces=donnees.drop_duplicates()
donnees_NaN_supprimes=donnees_NaN_supprimes.drop_duplicates()
print("Tableau avec NaN remplacés")
print(donnees_NaN_remplaces.shape)
print("Tableau avec suppression des lignes contenant des NaN")
print(donnees_NaN_supprimes.shape)
```

### 4.2.2. Graphique à barres groupées

```python
sports_10=donnees_NaN_remplaces['Sport'].value_counts()[:10].keys()
donnees_bar=donnees_NaN_remplaces.loc[donnees_NaN_remplaces['Sport'].isin(sports_10)].groupby(["Sport", "Sex"]).size().unstack(1)
donnees_bar
```

```python
import matplotlib.pyplot as plt
pos = list(range(len(donnees_bar['F']))) 
width = 0.25 
plt.figure()
plt.subplot(111)
# Créer une barre avec les données F,
# à la position pos
plt.bar(pos, donnees_bar['F'], width, color='red',label="Femmes") 
# Créer une barre avec les données M,
# à la position pos + 0.25 afin que la barre M soit décalée à droite de la première barre F
plt.bar([p + width for p in pos], donnees_bar['M'],width, color='blue', label="Hommes") 
plt.xticks([p + 0.125 for p in pos],donnees_bar.index,rotation=90)
plt.ylabel("Nombre d'athlètes")
plt.legend()
plt.title("Nombre d'athlètes hommes et femmes dans les 10 disciplines")
plt.show()
```

## 4.3. Box plots

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(111)
plt.boxplot(donnees_NaN_supprimes["Age"])
plt.show()
```

On constate beaucoup de valeurs extrêmes pour la variable âge. En effet, nous l’avions vu précédemment, certaines disciplines artistiques qui avaient lieu aux JO entre 1924 et 1948 regroupaient des athlètes très âgés, ce qui n’est pas représentatif de la majorité des athlètes des JO : en effet, ces disciplines ont très vite disparu, laissant la place à des disciplines nécessitant d’être plus jeune pour exceller. Cela explique ces valeurs extrêmes. 

On peut donc afficher notre boxplot sans les valeurs extrêmes, car elles aplatissent notre boîte à moustaches. 

Pour cela, nous mettrons l’option showfliers de la fonction boxplot() à False.

```python
import matplotlib.pyplot as plt
plt.figure()
plt.subplot(111)
plt.boxplot(donnees_NaN_supprimes["Age"],showfliers=False)
plt.title("Boite à moustache sur la variable Age")
plt.xticks([1], ['Age'])
plt.show()
```
